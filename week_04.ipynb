### list comprehension
- 포함형 / 지능형 / 축약형 리스트
- 하나의 리스트에서 그걸 포함하여 다른 리스트를 만든다.
- list + for loop
- [A for A' in B]
    - C에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
    - A : 내가 원하는 형태
    - A' : C의 형태
    - B : 반복 가능한 자료형 (리스트)
    
- 조건 추가 : [A for A' in B if ...]

2. text2에서 ch로 끝나는 단어를 출력하는데
    - 중복 없고
    - 첫 글자와 ch 남기고 가운데는 모두 -로 가려라.
        예 : much -> m-ch
        - ch로 시작하면 바꾸면 안됨
    - 정렬 : 
        - 문자 길이로 정렬하기

```python
# 1.
len_list = len([word for word in text6.tokens if (len(word) <= 4) and (word.isalpha()) and ('ie' in word)])
print(len_list)
```

```python
# 2.
text2_set = set([word for word in text2.tokens if word.endswith('ch')]) # 'ch'로 끝나는 단어들 중복없이 text2_set에 담음.
list = []
str = ""

for word in text2_set :
    if word.startswith('ch') : # 'ch'로 시작하면
        str = 'ch' + ''.join(['-' for char in word[2:-2]]) + 'ch' 
        # 'ch' + 3번째 글자 ~ 마지막에서 3번째 글자 '-'로 바꾼 리스트 합친 문자열 + 'ch'
        list.append(str) # 새로운 리스트에 집어 넣음.
    else :
        str = word[0] + ''.join(['-' for char in word[1:-2]]) + 'ch'
        # 첫 글자 + 2번째 글자 ~ 마지막에서 3번째 글자 '-'로 바꾼 리스트 합친 문자열 + 'ch'
        list.append(str) # 새로운 리스트에 집어 넣음.
        
text2_set = set(list) # '-'로 변환한 뒤 중복되는 것들 다시 삭제하기 위해 set로 변환
print(text2_set)
```
